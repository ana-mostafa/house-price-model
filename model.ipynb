```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("London.csv")
```

```python
data.head(100)
```

```python
data = data.drop(data.columns[0], axis=1)
data.head()
```

```python
mode_value = data['Location'].mode()[0]
data['Location'].fillna(mode_value, inplace=True)
```

```python
data.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the "Location" column
data['Location'] = label_encoder.fit_transform(data['Location'])
data['City/County'] = label_encoder.fit_transform(data['City/County'])
data['House Type'] = label_encoder.fit_transform(data['House Type'])
data.head()
```

```python
# Calculate the mean price for each unique area (postcode)
area_prices = data.groupby('Postal Code')['Price'].mean()

# Map the mean prices to the corresponding postcodes
data['Area Price'] = data['Postal Code'].map(area_prices)

# Drop the original 'Postal Code' column if necessary
data.drop('Postal Code', axis=1, inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['Price'], axis = 1)
y = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```
